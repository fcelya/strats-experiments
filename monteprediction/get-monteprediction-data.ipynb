```python
import pandas as pd
from monteprediction.api import get_mean, get_covariance
import requests
import pickle as pkl
import time
from monteprediction import SPDR_ETFS
import yfinance as yf
```

```python
def get_available_dates():
    url = "https://www.monteprediction.com/api/moments"
    
    try:
        # Make a GET request to the API
        response = requests.get(url)
        
        # Check if the request was successful
        if response.status_code == 200:
            data = response.json()  # Parse the JSON response
            return data.get("available_dates", [])  # Return the list of available dates
        else:
            print(f"Failed to retrieve data. Status code: {response.status_code}")
            return []
    
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Example usage
dates = get_available_dates()
print(dates)
```

```python
data = []
for date in dates:
    d = {}
    d['date'] = date
    mean = get_mean(date)
    d['mean'] = mean.to_numpy()[0]
    time.sleep(0.5)
    cov = get_covariance(date)
    d['cov'] = cov.to_numpy()
    d['assets'] = cov.columns.to_list()
    data.append(d)
    time.sleep(0.5)
with open('./data/predictions.pkl','wb') as f:
    pkl.dump(data,f)
```
