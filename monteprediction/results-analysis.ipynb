```python
import pandas as pd
import matplotlib.pyplot as plt
from quantstats.reports import full

DATA_FOLDER = './data/'
RESULTS_FOLDER = './results/'
```

```python
df_montepred_hrp = pd.read_csv(RESULTS_FOLDER+'montepred_hrp.csv',index_col=0,parse_dates=True)
df_equal_rebalance = pd.read_csv(RESULTS_FOLDER+'equal_rebalance.csv',index_col=0,parse_dates=True)
df_equal_constant = pd.read_csv(RESULTS_FOLDER+'equal_constant.csv',index_col=0,parse_dates=True)
df_montepred_mvo = pd.read_csv(RESULTS_FOLDER+'montepred_mvo.csv',index_col=0,parse_dates=True)
```

```python
plt.plot(df_montepred_hrp.index,df_montepred_hrp['initial_wealth'])
plt.plot(df_montepred_mvo.index,df_montepred_mvo['initial_wealth'])
plt.plot(df_equal_rebalance.index,df_equal_rebalance['initial_wealth'])
plt.plot(df_equal_constant.index,df_equal_constant['initial_wealth'])
plt.legend(['montepred-hrp','montepred-mvo','equal-rebalance','equal-constant'])
plt.tight_layout()
```
